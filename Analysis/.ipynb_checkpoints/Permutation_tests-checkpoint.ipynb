## Mood's median test for pairwise comparisons of gene expression between populations.
Compares equality of medians for two or more populations. Compared to Kruskal-Wallis test, and one-way ANOVA tests. 
Assumes that samples are independent, and that they have the same shape. Shapes can be tested using a probability plot. (https://www.isixsigma.com/tools-templates/hypothesis-testing/understanding-uses-moods-median-test/)

Generally, the Wilcoxon Rank Sum or Mann-Whitney test is used instead of Mood’s Median Test since they provide more accurate results. (https://www.real-statistics.com/non-parametric-tests/moods-median-test-two-samples/)

```python
# Test whether sample distributions have the same shape.
def prob_plot(df,tissue,gene,col_expr):
    data = df[(df['Tissue']==tissue)&(df['Gene']==gene)]
    data_euam = data[data['self_rep_race']=='EuAm']
    data_afam = data[data['self_rep_race']=='AfAm']
    # Histogram and QQplot (https://www.geeksforgeeks.org/normal-probability-plot/).
    fig, ax = plt.subplots(1, 4, figsize=(24, 7))
    # European Americans.
    sm.ProbPlot(data_euam[col_expr]).qqplot(line='s', ax=ax[1]);
    sns.histplot(data_euam[col_expr],kde=True, color ='blue',ax=ax[0]);
    # African Americans.
    sm.ProbPlot(data_afam[col_expr]).qqplot(line='s', ax=ax[3]);
    sns.histplot(data_afam[col_expr],kde=True, color ='red',ax=ax[2]);
    ax[0].set_title(gene+" for European Americans", size=20)
    ax[2].set_title(gene+" for African Americans", size=20)
    plt.show()
    
# Mood's median test: compares medians for two or more samples.
def moods_bonferroni( df, list_tissues,col_expr):
    out_list=[]
    for tissue in list_tissues:
        for gene in list_mtdna:
            data = df[(df['Gene']==gene)&(df['Tissue']==tissue)]
            data_euam = data[data['self_rep_race']=='EuAm']
            data_afam = data[data['self_rep_race']=='AfAm']
            # Mood's median test (similar to ANOVA and Kruskal-Wallis tests).
            mood = stats.median_test(data_afam[col_expr].tolist(),data_euam[col_expr].tolist())
            bonf_pvalue = mood[1]*len(list_mtdna)
            if bonf_pvalue > 1:
                bonf_pvalue = 1
            out_list.append([tissue,gene,mood[0],mood[1],bonf_pvalue,mood[2],mood[3]])
    return(pd.DataFrame(out_list, columns=['Tissue','Gene',"Mood's median test statistic",
                                           'P-value','Bonferroni p-value','Grand median','Contingency table']))
        

#[ prob_plot(df_adjTPM,'Muscle - Skeletal',gene,'Residuals_AgeIsch') for gene in list_mtdna ]
table_moods = moods_bonferroni(df_adjTPM,list_tissues,'Residuals_AgeIsch')
table_moods[table_moods['Bonferroni p-value']<=0.05]
```

## Mann- Whitney U test for pairwise comparisons of gene expression between populations.
(Non-parametric) two-sample rank test.

The Mann-Whitney U test is essentially an alternative form of the Wilcoxon Rank-Sum test for independent samples and is completely equivalent. Wilcoxon is for related/paired samples. (https://www.real-statistics.com/non-parametric-tests/mann-whitney-test/)

The only assumptions for carrying out a Mann-Whitney test are that the two groups must be independent and that the dependent variable is ordinal or numerical (continuous).

The MWU "Common Language Effect size" provided by pingouin package helps interpretability (https://www.tandfonline.com/doi/full/10.1177/0748175615578739).

```python
### Make table of the 'genes' compared across EuAm vs. AfAm, and 'p-values' of two-sided Mann-Whitney U test.
# Use only when the number of observation in each sample is > 20 and you have 2 independent samples of ranks. 
# Mann-Whitney U is significant if the u-obtained is LESS THAN or equal to the critical value of U.

# Mann-Whitney U test: Compares mean ranks of independent samples.
def mwu_bonferroni(df,tissue,col_expr):
    out_df= pd.DataFrame(columns=['Tissue','Gene','U-val','alternative','RBC','CLES','p-val','MedianExpr_AfAm','MedianExpr_EuAm','P-value (one-sided: Eu>Af)'])
    #col_tpm = col_expr
    for gene in list_mtdna:
        df_gene = df[(df['Gene']==gene)&(df['Tissue']==tissue)]
        # Split by population.
        data_euam = df_gene[df_gene['self_rep_race']=='EuAm']
        data_afam = df_gene[df_gene['self_rep_race']=='AfAm']
        ##return([ median(row) for row in data_afam[col_expr] ])
        
        # Mann-Whitney U test with Pingouin (https://pingouin-stats.org/generated/pingouin.mwu.html).
        # Ha: Expression in EuAm =/= to AfAm (two-sided).
        mwu_2s = pg.mwu(data_euam[col_expr], data_afam[col_expr], alternative='two-sided')
        # Ha: Expression in EuAm > than AfAm (one-sided).
        mwu_greater = pg.mwu(data_euam[col_expr], data_afam[col_expr], alternative='greater')
        #return(mwu_greater)
        
        ## Median expression for each population.
        mwu_2s['MedianExpr_AfAm'] = data_afam[col_expr].median()
        mwu_2s['MedianExpr_EuAm'] = data_euam[col_expr].median()
        # Add the one-sided p-value, tissue, and gene.
        mwu_2s['P-value (one-sided: Eu>Af)'] = mwu_greater['p-val']
        mwu_2s['Tissue'] = tissue
        mwu_2s['Gene'] = gene
        # Keep stats from one-sided test.
        mwu_2s['RBC'] = mwu_greater['RBC']
        mwu_2s['CLES'] = mwu_greater['CLES']
        
        # Export stats.
        out_df = pd.concat([out_df,mwu_2s])        
    
    # Drop the alternative.
    out_df = out_df.drop(columns=['alternative'])
    # Bonferroni-correction.    
    out_df['Bonferroni (one-sided: Eu>Af)'] = (out_df['P-value (one-sided: Eu>Af)']) * len(list_mtdna)
    out_df.loc[out_df['Bonferroni (one-sided: Eu>Af)'] >= 1, 'Bonferroni (one-sided: Eu>Af)'] = 1
    # Rename columns.
    out_df.columns = ['Tissue', 'Gene', 'U-value', 'RBC', 'CL Effect Size','P-value (two-sided)',
            'MedianExpr_AfAm','MedianExpr_EuAm','P-value (one-sided: Eu>Af)','Bonferroni (one-sided: Eu>Af)']
    return(out_df)


# MWU results for multiple tissues.
def multi_mwu_bonf(df,list_tissues,col_expr):
    out = pd.DataFrame(columns=['Tissue', 'Gene', 'U-value', 'RBC', 'CL Effect Size','P-value (two-sided)',
            'MedianExpr_AfAm','MedianExpr_EuAm','P-value (one-sided: Eu>Af)','Bonferroni (one-sided: Eu>Af)'])
    for tissue in list_tissues:
        out = out.append(mwu_bonferroni(df,tissue,col_expr))
    return(out)


def convert_pvalue_to_asterisks(pvalue):
    if pvalue <= 0.0001:
        return "****"
    elif pvalue <= 0.001:
        return "***"
    elif pvalue <= 0.01:
        return "**"
    elif pvalue <= 0.05:
        return "*"
    return "ns"


mwu_bonferroni(df_adjTPM,'Muscle - Skeletal','Residuals_AgeIsch')

table_mwu = multi_mwu_bonf(df_adjTPM,list_tissues,'Residuals_AgeIsch')
table_mwu#[table_mwu['Bonferroni (one-sided: Eu>Af)']<=0.05]
```

# Figure S6: Haplogroups H and L in AfAm and EuAm

```python
# Does haplogroup explain the difference in expression between groups?
def plot_pair_expr_haplo(df,tissue,col_expr):
    # Plot size and style.
    sns.set(rc={'figure.figsize':(30,15)},font_scale=2.5)
    sns.set_style("ticks")
    # Filter by tissue.
    data = df[df['Tissue']==tissue].sort_values('Gene')
    # Keep just H and L mt-haplogroups.
    data = data[data['mtDNA_haplo'].isin(['H','L'])]
    # Change name.
    data['Pop'] = [{'AfAm':'African Americans','EuAm':'European Americans'}[x] for x in data['self_rep_race']]
    # Create group labels: using ".loc" with multiple conditions.
    ##for pop in ['AfAm','EuAm']:
    for pop in data['Pop'].unique():
        for haplo in ['H','L']:
            # Use multiple conditions with .loc to assign label to group column.
            data.loc[(
                data['Pop']==pop) & (data['mtDNA_haplo']==haplo), #Multiple conditions.
                'Groups'] = haplo+' : '+pop  #Label.
    # Get list of groups and reorder it.
    groups = data['Groups'].unique()
    ## Reorderd by population (AfAm them EuAm).
    ##groups = [groups[1],groups[3],groups[2],groups[0]]
    # Reordered by haplogroup, then by matching population.
    groups = [groups[0],groups[3],groups[1],groups[2]]
    
    # Box plot of all genes for each group.
    sns.boxplot(data=data, x='Gene', y=col_expr, hue='Groups',
                hue_order=[*groups], palette="Oranges_r")
    plt.xticks(rotation=0, fontstyle='italic')
    plt.xlabel('\nGenes')
    if col_expr == 'TPM':
        plt.ylabel('Normalized gene expression (TPM)')
    elif col_expr == 'Residuals_AgeIsch':
        plt.ylabel('Adjusted gene expression')
    # Legend placement.
    plt.legend(loc="upper right")
    # Alternate tissue names.
    dict_altTiss = {'Esophagus - Muscularis':'Smooth muscle (esophagus)','Muscle - Skeletal':'Muscle (skeletal)','Heart - Left Ventricle':'Heart (ventricle)','Artery - Tibial':'Artery (Tibial)','Whole Blood':'Whole blood','Heart - Atrial Appendage':'Heart (atrium)','Nerve - Tibial':'Nerve (Tibial)'}
    # Sample sizes per group.
    ##groups = data['Groups'].unique()
    s_1 = str(groups[0])+"="+str(data[data['Groups']==groups[0]]['short_ID'].drop_duplicates().size)
    s_2 = groups[1]+" = "+str(data[data['Groups']==groups[1]]['short_ID'].drop_duplicates().size)
    s_3 = groups[2]+" = "+str(data[data['Groups']==groups[2]]['short_ID'].drop_duplicates().size)
    s_4 = groups[3]+" = "+str(data[data['Groups']==groups[3]]['short_ID'].drop_duplicates().size)
    plt.title( dict_altTiss[tissue] +"\n "+ s_1 +"\n"+ s_2 +"\n"+ s_3 +"\n"+ s_4+"\n", size=40, weight="bold" )
    ## Save figure.
    plt.savefig('../results/boxplot_skeletal_haplo_03282022.pdf')
    plt.show()
    # Get test results.
    mwu_L = mwu_pair_expr_haplo(data,tissue,col_expr, [groups[0],groups[1]] )
    mwu_H = mwu_pair_expr_haplo(data,tissue,col_expr, [groups[2],groups[3]] )
    #return(mwu_H)
    mwu = pd.concat([mwu_L,mwu_H])
    return(mwu)
    

# Test if expression for each mt-haplogroup is different between populations.
def mwu_pair_expr_haplo(df,tissue,col_expr,two_groups):
    out_df= pd.DataFrame(columns=['Comparison','Tissue','Gene','U-val','alternative','RBC','CLES','p-val','MedianExpr_1','MedianExpr_2','P-value (one-sided)'])
    #col_tpm = col_expr
    for gene in list_mtdna:
        df_gene = df[(df['Gene']==gene)&(df['Tissue']==tissue)]
        # Split by population.
        data_euam = df_gene[df_gene['Groups']==two_groups[0]]
        data_afam = df_gene[df_gene['Groups']==two_groups[1]]
        
        # Mann-Whitney U test with Pingouin (https://pingouin-stats.org/generated/pingouin.mwu.html).
        # Ha: Expression in EuAm =/= to AfAm (two-sided).
        mwu_2s = pg.mwu(data_euam[col_expr], data_afam[col_expr], alternative='two-sided')
        # Ha: Expression in EuAm > than AfAm (one-sided).
        mwu_greater = pg.mwu(data_euam[col_expr], data_afam[col_expr], alternative='greater')
        #return(mwu_greater)
        
        # Add the comparison.
        mwu_2s['Comparison'] = two_groups[0]+" and "+two_groups[1]
        # Median expression for each population.
        mwu_2s['MedianExpr_1'] = data_afam[col_expr].median()
        mwu_2s['MedianExpr_2'] = data_euam[col_expr].median()
        # Add the one-sided p-value, tissue, and gene.
        mwu_2s['P-value (one-sided)'] = mwu_greater['p-val']
        mwu_2s['Tissue'] = tissue
        mwu_2s['Gene'] = gene
        # Keep stats from one-sided test.
        mwu_2s['RBC'] = mwu_greater['RBC']
        mwu_2s['CLES'] = mwu_greater['CLES']
        
        # Export stats.
        out_df = pd.concat([out_df,mwu_2s])        
    
    # Drop the alternative.
    out_df = out_df.drop(columns=['alternative'])
    # Bonferroni-correction.    
    out_df['Bonferroni (one-sided)'] = (out_df['P-value (one-sided)']) * len(list_mtdna)
    out_df.loc[out_df['Bonferroni (one-sided)'] >= 1, 'Bonferroni (one-sided)'] = 1
    # Rename columns.
    out_df.columns = ['Comparison','Tissue', 'Gene', 'U-value', 'RBC', 'CL Effect Size','P-value (two-sided)',
            'MedianExpr_1','MedianExpr_2','P-value (one-sided)','Bonferroni (one-sided)']
    return(out_df)
    
    



#mwu_pair_expr_haplo()

plot_pair_expr_haplo(df_adjTPM,'Muscle - Skeletal','Residuals_AgeIsch')
```

# Permutation tests
- Statistic: difference in mean TPM below the MND cutoff, and mean TPM above the MND cutoff.
- Observed difference in mean TPM: Calculates this statistic in our observed data.
- Expected distribution of differences in mean TPM: For 1000 permutations, the statistic is calculated then plotted.
- one-sided P-value: The proportion of permutations with a result above the observed difference in mean TPM.

```python
# Permutation test.
def permut_mnd_tpm(df,tissue,gene,col_expr,cutoffMND):
    # Keep MND and expression for a single gene.
    data = df[(df['Tissue'] == tissue)&(df['Gene'] == gene)][['Gene','mitonucl_discord',col_expr]]
    if cutoffMND == 'median':
        cutoffMND = data['mitonucl_discord'].median()
    elif cutoffMND == 'mean':
        cutoffMND = data['mitonucl_discord'].mean()
    # Observed difference in mean expression.
    obs_diff = data[data['mitonucl_discord'] < cutoffMND][col_expr].mean() - data[data['mitonucl_discord'] >= cutoffMND][col_expr].mean()
    # Permutation values.
    perm_repl = 1000
    perm_results = []
    for seq in list(range(perm_repl)):
        # Set random seed.
        np.random.seed(40)
        # Permute MND labels.
        data['mitonucl_discord'] = np.random.permutation(data['mitonucl_discord'])
        # Split data at MND of 0.1, then get the mean expression (sample statistic).
        mean_up = data[data['mitonucl_discord'] >= cutoffMND][col_expr].mean()
        mean_down = data[data['mitonucl_discord'] < cutoffMND][col_expr].mean()
        # Add results.
        perm_results.append((mean_down,mean_up))
    # Add results into a dataframe (to be exported).
    results = pd.DataFrame(perm_results)
    results.columns = ['Below_MND_'+str(cutoffMND),'Above_MND_'+str(cutoffMND)]
    results['Difference'] = results['Below_MND_'+str(cutoffMND)] - results['Above_MND_'+str(cutoffMND)]
    # Estimate p-value.
    no_aboveObs = results[results['Difference'] > obs_diff].shape[0]
    pvalue = no_aboveObs/perm_repl
    # Bonferroni-corrected p-values.
    bonf_pvalue = pvalue*len(list_mtdna)
    # If Bonferroni-corrected p-value is above 1, reset to 1 (the upper range of all p-values).
    if bonf_pvalue > 1:
        bonf_pvalue = 1
    # P-value cannot actually be zero. With enough permutations this would become =/= 0.0
    elif bonf_pvalue == 0:
        bonf_pvalue = '<0.001'
    return(results,obs_diff,cutoffMND,pvalue,bonf_pvalue)


# Plot permutation results for each genes in a tissue.
def subplot_perm(df,tissue,list_mtdna,col_expr,cutoffMND):
    # Parameters for multiple plots / Subplotting.  (https://dev.to/thalesbruno/subplotting-with-matplotlib-and-seaborn-5ei8)
    fig, axes = plt.subplots(3,4, figsize=(40,30), sharex=False, sharey=True)
    num_plots=11
    perm_repl=1000
    # Choose ax_order.
    #ax_order = [axes[0, 0], axes[0, 1], axes[0, 2], axes[1, 0], axes[1, 1], axes[1, 2], axes[2, 0], axes[2, 1], axes[2, 2]]
    ax_order = [axes[0, 0], axes[0, 1], axes[0, 2], axes[0, 3], axes[1, 0], axes[1, 1], axes[1, 2], axes[1, 3], axes[2, 0], axes[2, 1], axes[2, 2], axes[2, 3]]
    #ax_order = [axes[0, 0], axes[0, 1], axes[0, 2], axes[0, 3], axes[1, 0], axes[1, 1], axes[1, 2], axes[1, 3], axes[2, 0], axes[2, 1], axes[2, 2], axes[3, 0], axes[3, 1]]
    # Plot results of permutation for each gene.
    for ( gene , ax_curr ) in zip( list_mtdna , ax_order ):  #(https://stackoverflow.com/questions/1663807/how-to-iterate-through-two-lists-in-parallel)
        # Run permutation.
        results,obs_diff,cutoffMND,pvalue,bonf_pvalue = permut_mnd_tpm(df,tissue,gene,col_expr,cutoffMND)
        # Plot results for this gene.
        sns.set(rc={'figure.figsize':(20,20)},font_scale=3.6)
        sns.set_style("ticks")
        sns.histplot( results['Difference'] , ax=ax_curr)
        ax_curr.set_xlim(-14000, 14000)
        ax_curr.set_title(gene, weight='bold', fontstyle='italic')
        ax_curr.set_ylabel("")
        ax_curr.set_xlabel("")
        # Plot line of the Observed mean difference in TPM .
        ax_curr.plot([obs_diff,obs_diff], [100, 0], linewidth=8, color='r')
        # Plot the p-value (the number of results above the observed value over the number of iterations of the test).
        no_aboveObs = results[results['Difference'] > obs_diff].shape[0]
        # Shorten bonferroni-corrected p-value if it's a float.
        if isinstance(bonf_pvalue, float) == True:
            bonf_pvalue = round(bonf_pvalue, 4)
        # Include p-value in subplot.
        if bonf_pvalue == '<0.001':
            ax_curr.text(2000, 115, 'p'+str(bonf_pvalue) , fontsize=40, color='black')
        else:
            ax_curr.text(2000, 115, 'p='+str(bonf_pvalue) , fontsize=40, color='black')
    # Title for the entire plot.
    fig.suptitle('Permutation test for '+tissue+' (MND cutoff: '+str(round(cutoffMND,3))+" )\nDistribution of expected differences in mean TPM between groups", 
                 size=55)
    # Manually insert y-axis and explain vertical red line.
    axes[1,0].set_ylabel('\nPermutations',
                         fontsize=60, ha='center')
    axes[1,1].set_xlabel('\n\n\n\n\n\n\n\n\n\nDifferences in mean expression between groups of high and low mitonuclear discordance' , 
                         fontsize=60, ha='center' )
    ##plt.text(-40000, -30,'Observed difference in mean TPM' , fontsize=60, color='r', ha='left')
    # Delete subplot.
    axes[2,3].remove()
    ## Save figure.
    #plt.savefig('../results/permutation_skeletal_adjusted_05172022.pdf')


def table_permut_mnd_tpm( df,list_tissue,col_expr,cutoffMND ):
    out = []
    # Iterate for each tissue and gene.
    for tissue in list_tissue:
        for gene in list_mtdna:
            # Get permutation results.
            results,obs_diff,cutoff,pvalue,bonf_pvalue = permut_mnd_tpm(df,tissue,gene,col_expr,cutoffMND)
            out.append([tissue,gene,cutoff,obs_diff,pvalue,bonf_pvalue])
    return(pd.DataFrame(out, columns=['Tissue','Gene','Cutoff_MND','Obs_diff','P-value','Bonferroni p-value']))



#permut_mnd_tpm(df_adjTPM,'Muscle - Skeletal','ATP6','Residuals_AgeIsch',0.1)[0]
#permut_mnd_tpm(df_adjTPM,'Muscle - Skeletal','ATP6','Residuals_AgeIsch','median')[0]
#subplot_perm(df_adjTPM,'Muscle - Skeletal',list_mtdna,'Residuals_AgeIsch','mean')
#table_permut_mnd_tpm(df_adjTPM,list_tissue,'Residuals_AgeIsch','mean')
```
