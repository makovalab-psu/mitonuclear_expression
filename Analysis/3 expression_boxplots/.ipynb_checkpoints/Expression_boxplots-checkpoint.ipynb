## `Function`: Pairwise boxplots of expression between populations, per gene and tissue.

```python
# Pairwise boxplots of expression between populations (Figure 2A).
def plot_pair_expr(df,tissue,col_expr):
    sns.set(rc={'figure.figsize':(30,15)},font_scale=2.5)
    sns.set_style("ticks")
    data = df[df['Tissue']==tissue].sort_values('Gene')
    # Change labels of Self_rep_race column.
    data['Groups'] = data['self_rep_race']
    data['Groups'] = [{'AfAm':'African Americans','EuAm':'European Americans'}[x] for x in data['Groups']]
    # Pairwise boxplots.
    sns.boxplot(data=data, x='Gene', y=col_expr, hue='Groups', 
                hue_order=['African Americans','European Americans'],palette="Blues_r")
    # Plot options.
    plt.xticks(rotation=0, fontstyle='italic')
    plt.xlabel('\nGenes')
    if col_expr == 'TPM':
        plt.ylabel('Normalized gene expression (TPM)')
    elif col_expr == 'Residuals_AgeIsch':
        plt.ylabel('Adjusted gene expression')
    # Legend placement.
    plt.legend(loc="upper right")
    # Alternate tissue names.
    dict_altTiss = {'Esophagus - Muscularis':'Smooth muscle (esophagus)','Muscle - Skeletal':'Muscle (skeletal)','Heart - Left Ventricle':'Heart (ventricle)','Artery - Tibial':'Artery (Tibial)','Whole Blood':'Whole blood','Heart - Atrial Appendage':'Heart (atrium)','Nerve - Tibial':'Nerve (Tibial)'}
    # Sample sizes of populations.
    s_ea = "EuAm="+str(data[data['self_rep_race']=='EuAm']['short_ID'].drop_duplicates().size)
    s_aa = "AfAm="+str(data[data['self_rep_race']=='AfAm']['short_ID'].drop_duplicates().size)
    plt.title( dict_altTiss[tissue] +"( "+ s_ea +" , "+ s_aa +" )\n", size=40, weight="bold" )
    ## Save figure.
    #plt.savefig('../results/boxplot_skeletal_adjTPM_05172022.pdf')
    plt.show()


# Pairwise boxplots for multiple genes (Figure S4).
def subplot_pair_expr(df,list_tissue,col_expr):
    # Seaborn parameters.
    sns.set(rc={'figure.figsize':(15,15)},font_scale=3)
    sns.set_style("ticks")
    # Multiple plots / Subplotting.  (https://dev.to/thalesbruno/subplotting-with-matplotlib-and-seaborn-5ei8)
    fig, axes = plt.subplots(3,2, figsize=(40,60), sharex=False, sharey=False)
    num_plots=6
    # Iterate through all axes.  (https://stackoverflow.com/questions/41329789/populating-seaborn-subplots-using-an-array)
    ax_order=[]
    for i in range(num_plots):
        row = i % 3
        col = i // 3
        ax_curr = axes[row, col]
        ax_order.append(ax_curr)
    # Plot results for each gene.
    for (tissue,ax_curr) in zip(list_tissue,ax_order):
        # Dataset.
        data = df[df['Tissue']==tissue].sort_values('Gene')
        # Change labels of Self_rep_race column.
        data['Groups'] = data['self_rep_race']
        data['Groups'] = [{'AfAm':'African Americans','EuAm':'European Americans'}[x] for x in data['Groups']]
        # Plot.
        sns.boxplot(data=data, x='Gene', y=col_expr, hue='Groups', palette="Blues_r",
                    hue_order=['African Americans','European Americans'], ax=ax_curr)
        ax_curr.tick_params('x', labelrotation=45)
        ax_curr.set_xlabel('')
        ax_curr.set_ylabel('')
        # Legend placement.
        ax_curr.legend(loc="upper right")
        # Sample sizes.
        s_ea = "EuAm="+str(data[data['self_rep_race']=='EuAm']['short_ID'].drop_duplicates().size)
        s_aa = "AfAm="+str(data[data['self_rep_race']=='AfAm']['short_ID'].drop_duplicates().size)
        # Alternate tissue names.
        dict_altTiss = {'Esophagus - Muscularis':'Smooth muscle (esophagus)','Muscle - Skeletal':'Skeletal muscle','Heart - Left Ventricle':'Heart muscle (ventricle)','Artery - Tibial':'Smooth muscle (tibial artery)','Whole Blood':'Whole blood','Heart - Atrial Appendage':'Heart muscle (atrium)','Nerve - Tibial':'Nerve (tibial)'}
        # Subplot title.
        ax_curr.set_title( dict_altTiss[tissue] +"\n( "+ s_ea +" , "+ s_aa +" )", size=40, weight="bold" )
    # Set only one x and y label for the entire plot.
    if col_expr == 'TPM':
        axes[1,0].set_ylabel("             Normalized gene expression (TPM)\n", size=70) 
    elif col_expr == 'Residuals_AgeIsch':
        axes[1,0].set_ylabel("             Adjusted gene expression\n", size=70) 
    # X label.
    axes[2,0].set_xlabel("\n                                                   Genes", size=70) 
    ## Save figure.
    #plt.savefig('../results/boxplot_alltissues_adjTPM_05172022.pdf')

    
# Test whether each boxplot approximates a normal distribution (requirement for t-test of means).
def test_normality( df, tissue ):
    # Iterate for each gene.
    for gene in list_mtdna:
        df_gene = df[(df['Gene']==gene)&(df['Tissue']==tissue)]
        # Iterate for each population.
        for pop in ['EuAm','AfAm']:
            data_pop = df_gene[df_gene['self_rep_race']== pop ]        
            # Normality test.
            k2, p = stats.normaltest(data_pop['TPM'], axis=0, nan_policy='propagate')
            # Alpha value.
            alpha = 0.05
            # Interpretation of result.
            if p < alpha:  # null hypothesis: x comes from a normal distribution
                print("The null hypothesis can be rejected for " + gene + " in " + pop)
            else:
                print("The null hypothesis CANNOT be rejected for " + gene + " in " + pop)
    

# Change Pandas settings: number of rows
pd.set_option('display.max_rows', 100)

#plot_pair_expr(df_adjTPM,'Muscle - Skeletal')
#test_normality(df_adjTPM, 'Muscle - Skeletal')
```

## Mann- Whitney U test for pairwise comparisons of gene expression between populations.
(Non-parametric) two-sample rank test.

The Mann-Whitney U test is essentially an alternative form of the Wilcoxon Rank-Sum test for independent samples and is completely equivalent. Wilcoxon is for related/paired samples. (https://www.real-statistics.com/non-parametric-tests/mann-whitney-test/)

The only assumptions for carrying out a Mann-Whitney test are that the two groups must be independent and that the dependent variable is ordinal or numerical (continuous).

The MWU "Common Language Effect size" provided by pingouin package helps interpretability (https://www.tandfonline.com/doi/full/10.1177/0748175615578739).

```python
### Make table of the 'genes' compared across EuAm vs. AfAm, and 'p-values' of two-sided Mann-Whitney U test.
# Use only when the number of observation in each sample is > 20 and you have 2 independent samples of ranks. 
# Mann-Whitney U is significant if the u-obtained is LESS THAN or equal to the critical value of U.

# Mann-Whitney U test: Compares mean ranks of independent samples.
def mwu_bonferroni(df,tissue,col_expr):
    out_df= pd.DataFrame(columns=['Tissue','Gene','U-val','alternative','RBC','CLES','p-val','MedianExpr_AfAm','MedianExpr_EuAm','P-value (one-sided: Eu>Af)'])
    #col_tpm = col_expr
    for gene in list_mtdna:
        df_gene = df[(df['Gene']==gene)&(df['Tissue']==tissue)]
        # Split by population.
        data_euam = df_gene[df_gene['self_rep_race']=='EuAm']
        data_afam = df_gene[df_gene['self_rep_race']=='AfAm']
        ##return([ median(row) for row in data_afam[col_expr] ])
        
        # Mann-Whitney U test with Pingouin (https://pingouin-stats.org/generated/pingouin.mwu.html).
        # Ha: Expression in EuAm =/= to AfAm (two-sided).
        mwu_2s = pg.mwu(data_euam[col_expr], data_afam[col_expr], alternative='two-sided')
        # Ha: Expression in EuAm > than AfAm (one-sided).
        mwu_greater = pg.mwu(data_euam[col_expr], data_afam[col_expr], alternative='greater')
        #return(mwu_greater)
        
        ## Median expression for each population.
        mwu_2s['MedianExpr_AfAm'] = data_afam[col_expr].median()
        mwu_2s['MedianExpr_EuAm'] = data_euam[col_expr].median()
        # Add the one-sided p-value, tissue, and gene.
        mwu_2s['P-value (one-sided: Eu>Af)'] = mwu_greater['p-val']
        mwu_2s['Tissue'] = tissue
        mwu_2s['Gene'] = gene
        # Keep stats from one-sided test.
        mwu_2s['RBC'] = mwu_greater['RBC']
        mwu_2s['CLES'] = mwu_greater['CLES']
        
        # Export stats.
        out_df = pd.concat([out_df,mwu_2s])        
    
    # Drop the alternative.
    out_df = out_df.drop(columns=['alternative'])
    # Bonferroni-correction.    
    out_df['Bonferroni (one-sided: Eu>Af)'] = (out_df['P-value (one-sided: Eu>Af)']) * len(list_mtdna)
    out_df.loc[out_df['Bonferroni (one-sided: Eu>Af)'] >= 1, 'Bonferroni (one-sided: Eu>Af)'] = 1
    # Rename columns.
    out_df.columns = ['Tissue', 'Gene', 'U-value', 'RBC', 'CL Effect Size','P-value (two-sided)',
            'MedianExpr_AfAm','MedianExpr_EuAm','P-value (one-sided: Eu>Af)','Bonferroni (one-sided: Eu>Af)']
    return(out_df)


# MWU results for multiple tissues.
def multi_mwu_bonf(df,list_tissues,col_expr):
    out = pd.DataFrame(columns=['Tissue', 'Gene', 'U-value', 'RBC', 'CL Effect Size','P-value (two-sided)',
            'MedianExpr_AfAm','MedianExpr_EuAm','P-value (one-sided: Eu>Af)','Bonferroni (one-sided: Eu>Af)'])
    for tissue in list_tissues:
        out = out.append(mwu_bonferroni(df,tissue,col_expr))
    return(out)


def convert_pvalue_to_asterisks(pvalue):
    if pvalue <= 0.0001:
        return "****"
    elif pvalue <= 0.001:
        return "***"
    elif pvalue <= 0.01:
        return "**"
    elif pvalue <= 0.05:
        return "*"
    return "ns"


mwu_bonferroni(df_adjTPM,'Muscle - Skeletal','Residuals_AgeIsch')

table_mwu = multi_mwu_bonf(df_adjTPM,list_tissues,'Residuals_AgeIsch')
table_mwu#[table_mwu['Bonferroni (one-sided: Eu>Af)']<=0.05]
```

# Figure S6: Haplogroups H and L in AfAm and EuAm

```python
# Does haplogroup explain the difference in expression between groups?
def plot_pair_expr_haplo(df,tissue,col_expr):
    # Plot size and style.
    sns.set(rc={'figure.figsize':(30,15)},font_scale=2.5)
    sns.set_style("ticks")
    # Filter by tissue.
    data = df[df['Tissue']==tissue].sort_values('Gene')
    # Keep just H and L mt-haplogroups.
    data = data[data['mtDNA_haplo'].isin(['H','L'])]
    # Change name.
    data['Pop'] = [{'AfAm':'African Americans','EuAm':'European Americans'}[x] for x in data['self_rep_race']]
    # Create group labels: using ".loc" with multiple conditions.
    ##for pop in ['AfAm','EuAm']:
    for pop in data['Pop'].unique():
        for haplo in ['H','L']:
            # Use multiple conditions with .loc to assign label to group column.
            data.loc[(
                data['Pop']==pop) & (data['mtDNA_haplo']==haplo), #Multiple conditions.
                'Groups'] = haplo+' : '+pop  #Label.
    # Get list of groups and reorder it.
    groups = data['Groups'].unique()
    ## Reorderd by population (AfAm them EuAm).
    ##groups = [groups[1],groups[3],groups[2],groups[0]]
    # Reordered by haplogroup, then by matching population.
    groups = [groups[0],groups[3],groups[1],groups[2]]
    
    # Box plot of all genes for each group.
    sns.boxplot(data=data, x='Gene', y=col_expr, hue='Groups',
                hue_order=[*groups], palette="Oranges_r")
    plt.xticks(rotation=0, fontstyle='italic')
    plt.xlabel('\nGenes')
    if col_expr == 'TPM':
        plt.ylabel('Normalized gene expression (TPM)')
    elif col_expr == 'Residuals_AgeIsch':
        plt.ylabel('Adjusted gene expression')
    # Legend placement.
    plt.legend(loc="upper right")
    # Alternate tissue names.
    dict_altTiss = {'Esophagus - Muscularis':'Smooth muscle (esophagus)','Muscle - Skeletal':'Muscle (skeletal)','Heart - Left Ventricle':'Heart (ventricle)','Artery - Tibial':'Artery (Tibial)','Whole Blood':'Whole blood','Heart - Atrial Appendage':'Heart (atrium)','Nerve - Tibial':'Nerve (Tibial)'}
    # Sample sizes per group.
    ##groups = data['Groups'].unique()
    s_1 = str(groups[0])+"="+str(data[data['Groups']==groups[0]]['short_ID'].drop_duplicates().size)
    s_2 = groups[1]+" = "+str(data[data['Groups']==groups[1]]['short_ID'].drop_duplicates().size)
    s_3 = groups[2]+" = "+str(data[data['Groups']==groups[2]]['short_ID'].drop_duplicates().size)
    s_4 = groups[3]+" = "+str(data[data['Groups']==groups[3]]['short_ID'].drop_duplicates().size)
    plt.title( dict_altTiss[tissue] +"\n "+ s_1 +"\n"+ s_2 +"\n"+ s_3 +"\n"+ s_4+"\n", size=40, weight="bold" )
    ## Save figure.
    plt.savefig('../results/boxplot_skeletal_haplo_03282022.pdf')
    plt.show()
    # Get test results.
    mwu_L = mwu_pair_expr_haplo(data,tissue,col_expr, [groups[0],groups[1]] )
    mwu_H = mwu_pair_expr_haplo(data,tissue,col_expr, [groups[2],groups[3]] )
    #return(mwu_H)
    mwu = pd.concat([mwu_L,mwu_H])
    return(mwu)
    

# Test if expression for each mt-haplogroup is different between populations.
def mwu_pair_expr_haplo(df,tissue,col_expr,two_groups):
    out_df= pd.DataFrame(columns=['Comparison','Tissue','Gene','U-val','alternative','RBC','CLES','p-val','MedianExpr_1','MedianExpr_2','P-value (one-sided)'])
    #col_tpm = col_expr
    for gene in list_mtdna:
        df_gene = df[(df['Gene']==gene)&(df['Tissue']==tissue)]
        # Split by population.
        data_euam = df_gene[df_gene['Groups']==two_groups[0]]
        data_afam = df_gene[df_gene['Groups']==two_groups[1]]
        
        # Mann-Whitney U test with Pingouin (https://pingouin-stats.org/generated/pingouin.mwu.html).
        # Ha: Expression in EuAm =/= to AfAm (two-sided).
        mwu_2s = pg.mwu(data_euam[col_expr], data_afam[col_expr], alternative='two-sided')
        # Ha: Expression in EuAm > than AfAm (one-sided).
        mwu_greater = pg.mwu(data_euam[col_expr], data_afam[col_expr], alternative='greater')
        #return(mwu_greater)
        
        # Add the comparison.
        mwu_2s['Comparison'] = two_groups[0]+" and "+two_groups[1]
        # Median expression for each population.
        mwu_2s['MedianExpr_1'] = data_afam[col_expr].median()
        mwu_2s['MedianExpr_2'] = data_euam[col_expr].median()
        # Add the one-sided p-value, tissue, and gene.
        mwu_2s['P-value (one-sided)'] = mwu_greater['p-val']
        mwu_2s['Tissue'] = tissue
        mwu_2s['Gene'] = gene
        # Keep stats from one-sided test.
        mwu_2s['RBC'] = mwu_greater['RBC']
        mwu_2s['CLES'] = mwu_greater['CLES']
        
        # Export stats.
        out_df = pd.concat([out_df,mwu_2s])        
    
    # Drop the alternative.
    out_df = out_df.drop(columns=['alternative'])
    # Bonferroni-correction.    
    out_df['Bonferroni (one-sided)'] = (out_df['P-value (one-sided)']) * len(list_mtdna)
    out_df.loc[out_df['Bonferroni (one-sided)'] >= 1, 'Bonferroni (one-sided)'] = 1
    # Rename columns.
    out_df.columns = ['Comparison','Tissue', 'Gene', 'U-value', 'RBC', 'CL Effect Size','P-value (two-sided)',
            'MedianExpr_1','MedianExpr_2','P-value (one-sided)','Bonferroni (one-sided)']
    return(out_df)
    
    



#mwu_pair_expr_haplo()

plot_pair_expr_haplo(df_adjTPM,'Muscle - Skeletal','Residuals_AgeIsch')
```
